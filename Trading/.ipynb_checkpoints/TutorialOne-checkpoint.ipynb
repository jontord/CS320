```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
```

```python
style.use('ggplot')

start = dt.datetime(2018, 12, 1)
end = dt.datetime.now()
```

```python
df = web.DataReader("AAPL", 'yahoo', start, end)
df.head()
```

```python
df2 = web.DataReader("MSFT", "yahoo", start, end)
df2.head()
```

```python
#I want to graph the growth in microsoft stock
df2.tail()
```

```python
MSFTHighs = df2['High'].tolist()
plt.plot(MSFTHighs, color="blue")
plt.axis([0, 230, 95, 160])
plt.xlabel('time (days)')
plt.ylabel('value (dollars)')
plt.title("Microsoft Stock")
plt.show()
```

```python
begin = dt.datetime(1996, 1, 1)
close = dt.datetime(2019, 1, 1)

df3 = web.DataReader('AAPL', 'yahoo', begin, close)
df3.tail()
```

```python
appleStock = df3['High'].tolist()
plt.plot(appleStock, color="red")
plt.title("Apple Stock")
plt.show()
```
